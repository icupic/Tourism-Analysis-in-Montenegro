```python
import numpy as np
import pandas as pd
```

```python
tourists_by_country = pd.read_excel(r'C:\Users\Ivan\Desktop\tourism_analysis\data\2022\ukupno_2022.xlsx', sheet_name='Sheet3', header=1)
```

tourists_by_country - Tabela 3 Dolasci i noćenja stranih turista po zemlji pripadnosti, 2022. godina		

```python
tourists_by_country.rename(
    columns={'Zemlja pripadnosti gostiju': 'zemlja_porijekla', 
             'Dolasci': 'dolasci',
             'Struktura, u %': 'dolasci_u_%',
             'Noćenja':'nocenja',
             'Struktura , u %': 'nocenja_u_%'}, inplace= True)
```

Removed summed rows - like Europe, Total ect.

```python
filt = (~tourists_by_country['zemlja_porijekla'].isin(['Strani turisti', 'Evropa', 'Vanevropske zemlje']))
tourists_by_country = tourists_by_country[filt]
```

```python
tourists_by_country = tourists_by_country.drop(columns=['dolasci_u_%', 'nocenja_u_%'])
```

Druga tabla dolasci i nocenja po opstinama

```python
tourists_by_municipality = pd.read_excel(
    r'C:\Users\Ivan\Desktop\tourism_analysis\data\2022\ukupno_2022.xlsx', 
    sheet_name='Sheet2',
    header=3)
```

renameovanje kolona kao i brisanje vec sumiranih podataka

```python
tourists_by_municipality.rename(columns={
        'Unnamed: 0': 'opstina',
        '(1)': 'dolasci_stranih_turista',
        '(2)': 'dolasci_domacih_turista',
        '(1).1': 'nocenje_stranih_turista',
        '(2).1': 'nocenje_domacih_turista'},
        inplace=True
)
```

```python
tourists_by_municipality.drop(labels=0, inplace=True)
```

```python
tourists_by_municipality.drop(
    columns=['(3)=(1)+(2)', '(3) u %', '(3)=(1)+(2).1', '(3) u %.1'],
    inplace=True)
```

Treca tabela - Dolasci i nocenja turista po vrstama smjestaja

```python
tourists_by_accomodation = pd.read_excel(
    r'C:\Users\Ivan\Desktop\tourism_analysis\data\2022\ukupno_2022.xlsx', 
    sheet_name='Sheet1',
    header=6)
tourists_by_accomodation.head()
```

```python
tourists_by_accomodation.rename(columns={
        'Objekti kolektivnog smještaja': 'vrsta_smjestaja',
        1025642: 'dolasci_stranih_turista',
        140333: 'dolasci_domacih_turista',
        3787237: 'nocenje_stranih_turista',
        528457: 'nocenje_domacih_turista'},
        inplace=True
        
)
```

```python
tourists_by_accomodation.head()
```

```python
tourists_by_accomodation.drop(
    columns=[1165975, 4315694],
    inplace=True
)
```

```python
tourists_by_accomodation = tourists_by_accomodation[~tourists_by_accomodation['vrsta_smjestaja'].isnull()]
tourists_by_accomodation
```

```python
tourists_by_accomodation[[
    'dolasci_stranih_turista', 'dolasci_domacih_turista','nocenje_stranih_turista','nocenje_domacih_turista']] = 
tourists_by_accomodation[[
    'dolasci_stranih_turista', 
    'dolasci_domacih_turista',
    'nocenje_stranih_turista', 
    'nocenje_domacih_turista']].astype('int64')
```

```python
tourists_by_accomodation.loc[:,tourists_by_accomodation.dtypes == 'float64'] = tourists_by_accomodation.loc[:,tourists_by_accomodation.dtypes == 'float64'].astype('int64')
```
